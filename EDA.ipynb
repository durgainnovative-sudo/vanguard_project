```python
import pandas as pd
#Reading datasets
df_merged =pd.read_csv("df_merged.csv")
df_merged['date_time'] = pd.to_datetime(df_merged['date_time'])
df_merged.head()
```

```python
df_merged["variation"]=df_merged["variation"].fillna("results not included")
df_merged["clnt_tenure_yr"]=df_merged["clnt_tenure_yr"].fillna("unknown")
df_merged["clnt_tenure_mnth"]=df_merged["clnt_tenure_mnth"].fillna("unknown")
df_merged["clnt_age"]=df_merged["clnt_age"].fillna("unknown")
df_merged["gendr"]=df_merged["gendr"].fillna("unknown")
df_merged["bal"]=df_merged["bal"].fillna("unknown")
df_merged["calls_6_mnth"]=df_merged["calls_6_mnth"].fillna("unknown")
df_merged["logons_6_mnth"]=df_merged["logons_6_mnth"].fillna("unknown")
df_merged["num_accts"]=df_merged["variation"].fillna("unknown")

```

```python
df_merged["gendr"]=df_merged["gendr"].replace("X","unknown")
df_merged["gendr"]=df_merged["gendr"].replace("U","unknown")
```
